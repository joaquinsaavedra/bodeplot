```python
from pylab import *
        
```

```python
def H(w):
    alfa = 7.092
    beta = 2200000
    wc = 4000*pi
    s=1j*w
    return 150 /((1+s/alfa)*(1+s/beta))
```

```python
f = logspace(-1,6.8) # rango de frecuencias desde 10^-1 a 10^7
```

```python
plot(f, 20*log10(abs(H(2*pi*f)))); xscale('log')
plt.xlabel ("Frecuencia (Hz)")
plt.ylabel ("Amplitud (dB)")
plt.grid(True, which="both")
figure(figsize=(10,10),dpi=300)
```
